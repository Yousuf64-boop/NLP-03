```python
from gensim.models import word2vec
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')
```

```python
text = [
    "I love data science",
    "I love machine learning",
    "data science is amazing",
    "machine learning is powerful"
]
```

```python
nltk.download('punkt_tab')
tokenized_data = [word_tokenize(sentence.lower()) for sentence in text]
print("tokenized_data:\n",tokenized_data)
```

```python
model = word2vec.Word2Vec(sentences=tokenized_data,
                 vector_size=100,
                 window=2,
                 min_count=1,
                )
```

```python
vector = model.wv['data']
print(vector)
print("Vector length:", len(vector))
```

```python
similar_words = model.wv.most_similar('data')
print(similar_words)
```

```python
similarity = model.wv.similarity('data', 'science')
print("Similarity:", similarity)
```

```python
print(model.wv.index_to_key)
```

```python
model.save('word2vec.model')
model = word2vec.Word2Vec.load('word2vec.model')
```
